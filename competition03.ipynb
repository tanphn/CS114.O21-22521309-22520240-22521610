### Bước 2 - Data Augmentation

- Làm giàu dữ liệu bằng các công cụ có sẵn từ các framework như Keras



```python
# Create ImageDataGenerator instance with desired augmentations
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
)
```

```python
def plot_augmented_images(datagen, img_path, n_augmented=4, nImgHeight=150, nImgWidth=150):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_original = cv2.resize(img, (nImgHeight, nImgWidth))  # Resize original image
    img = np.expand_dims(img_original, axis=0)

    fig, axes = plt.subplots(1, n_augmented + 1, figsize=(20, 8))  # 1 row with original and augmented images
    axes[0].imshow(img_original)
    axes[0].set_title('Original')
    axes[0].axis('off')

    for i, ax in enumerate(axes[1:], start=1):
        augmented_img = datagen.flow(img, batch_size=1)[0][0].astype('uint8')
        ax.imshow(augmented_img)
        ax.set_title(f'Augmented {i}')
        ax.axis('off')

    fig.tight_layout()
    plt.show()
```

### Bước 3 - Feature Extraction

Từng bước:

Bước 3.1: Phát hiện xe máy (Dùng YOLOv5)

Bước 3.2: Cắt ảnh theo bounding box (Chỉ lấy ảnh có bounding box lớn nhất)

Bước 3.3: Trích xuất đặc trưng (Dùng MobileNet)

```python
# Khởi tạo model dùng để detect
model_detect = torch.hub.load('ultralytics/yolov5', 'yolov5s')
# Khởi tạo model dùng để trích xuất đặc trưng
mobilenet = models.mobilenet_v2(pretrained=True)
mobilenet.eval() # Chuyển mô hình sang chế độ đánh giá
# Chuẩn bị bộ chuyển đổi ảnh về tensor
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
# Định nghĩa hàm phát hiện, crop và trích xuất đặc trưng ảnh
def detectAndCropAndExFeature(model_detect, model_extract_feature, preprocess, img_path):
  img = Image.open(img_path)

  # Phát hiện đối tượng
  results = model_detect(img)

  # Lấy thông tin bounding box
  bounding_boxes = results.xyxy[0].numpy()

  # Lấy bounding boxes cho xe máy
  motorcycle_boxes = [box for box in bounding_boxes if int(box[5]) == 3]

  # Crop các vùng ảnh chứa xe máy
  max_area = 0 # Biến lưu diện tích lớn nhất của ảnh crop
  max_area_img = None # Biến lưu hình ảnh lớn nhất
  for i, box in enumerate(motorcycle_boxes):
      x1, y1, x2, y2 = map(int, box[:4]) # (x1, y1) là tọa độ của điểm góc trên bên trái, (x2, y2) là tọa độ của điểm góc dưới bên phải
      if (abs(x2 - x1) * abs(y2 - y1)) > max_area:
        max_area = abs(x2 - x1) * abs(y2 - y1)
        max_area_img = img.crop((x1, y1, x2, y2)) # Chỉ lấy ảnh crop có size lớn nhất
  if max_area_img is None:
    return np.empty((0,))
  img_tensor = preprocess(max_area_img).unsqueeze(0)  # Thêm batch dimension
  with torch.no_grad(): # Tắt tính năng tự động tính toán gradient (do không phải đang training mô hình nên không cần tính gradient)
      features = model_extract_feature.features(img_tensor)

  return np.array(features)
```

```python
file_path_train = szWorkingDir +'/'+ szTrainFileName

# Đọc file CSV vào DataFrame với header=None và đặt tên cho các cột
df = pd.read_csv(file_path_train, header=None, names=['ImageFullPath', 'CategoryID'])
df = df.head(10)
file_path_train
```

```python
def extract_features(df, dataset_path: str, outputdir: str = '/content/drive/MyDrive/feature_data', demo = True):
    features_list = [] #Danh sách lưu trữ feature của từng ảnh
    category_id_list = [] #Danh sách lưu trữ label của từng ảnh
    image_paths = [] #Danh sách lưu trữ đường dẫn của từng ảnh
    i = 0
    # Chuyển đổi đường dẫn dataset thành Path object
    dataset_path = Path(dataset_path)

    # Duyệt qua tất cả các dòng trong dataframe
    for index, row in df.iterrows():
        img_path = dataset_path / row['ImageFullPath']
        category_id = row['CategoryID']
        i += 1
        print(f'- Ảnh thứ {i}' )
        try:
            features = detectAndCropAndExFeature(model_detect, mobilenet, preprocess, img_path)
            if len(features) == 0:
              print('Không detect được xe')
              continue
            # In ra số chiều của vector đặc trưng
            print(f"Số chiều của vector đặc trưng: {features.shape}")

            features_list.append(features.flatten())
            category_id_list.append(category_id)
            image_paths.append(str(img_path))
        except Exception as e:
            print(f"Lỗi khi xử lý ảnh {img_path}: {e}")
        if demo and i == 20:
            break
    return image_paths, features_list, category_id_list

```
