**DỰ ĐOÁN ĐIỂM QUÁ TRÌNH**

```python
#Định dạng 'datatime' cho cột created_at, updated_at
for fmt in ['%m-%d %H:%M:%S', '%d-%m %H:%M:%S', '%d-%m-%Y %H:%M:%S']:
    try:
        annonimized_df['created_at'] = pd.to_datetime(annonimized_df['created_at'], format=fmt, errors='coerce')
        annonimized_df['updated_at'] = pd.to_datetime(annonimized_df['updated_at'], format=fmt, errors='coerce')
        if annonimized_df['created_at'].isnull().sum() == 0 and annonimized_df['updated_at'].isnull().sum() == 0:
            break
    except Exception as e:
        print(f"Format {fmt} không phù hợp: {e}")
```

```python
# Bảng gồm: assignment_id, số lượng problem trong assignment, thời gian bắt đầu, thời gian kết thúc
assignment_df = annonimized_df.groupby('assignment_id').agg({
                'problem_id' : 'nunique', # Số lượng bài trong từng assignment
                'created_at' : 'min', # Coi như thời gian tạo bài nhỏ nhất cũng là lúc assignment bắt đầu
                'updated_at' : 'max' # Coi như thời gian updated lớn nhất cũng là lúc assignment kết thúc
            }).reset_index()
```

```python
assignment_df.columns = ['assignment_id', 'problem', 'begin at', 'end at']
```

```python
assignment_df
```

```python
#  Bảng gồm: user name, problem_id,
#            điểm max của problem đó,
#            thời gian bắt đầu làm, số lần submit)
username_problem_df = annonimized_df.groupby(['username', 'assignment_id', 'problem_id']).agg({
                'pre_score' : 'max',
                'created_at' : 'min',
                'is_final' : 'count'
            }).reset_index()
```

```python
username_problem_df.columns = ['username', 'assignment_id', 'problem_id', 'max_score', 'create at', 'num of submit']
```

```python
# Bảng gồm: username, assignment_id mà user đó tham gia,
#           Số lượng problem mà user tham gia trong assignment đó,
#           Số lượng problem mà user hoàn thành trong assignment đó
username_assignment_df = annonimized_df.groupby(['username', 'assignment_id']).agg({
                'problem_id' : 'nunique',
                'is_final' : 'sum'
            }).reset_index()
```

```python
#Đổi tên cột cho username_assignment_df
username_assignment_df.columns = ['username', 'assignment_id', 'attend', 'finish']
```

```python
#Bảng gồm: username, assigment_id, giá trị nhỏ nhất của coefficient
#Mục đích: Liệt kê ra username, assigment_id, giá trị nhỏ nhất của coefficient nếu giá trị coefficient nhỏ hơn 100 chứng tỏ rằng user này có nộp trễ bài trong assignment đó
username_assignment_df_late = annonimized_df.groupby(['username', 'assignment_id']).agg({
                'coefficient' : 'min',
            }).reset_index()
#Bảng gồm: username, tổng coefficient
username_sumCoef = username_assignment_df_late.groupby('username').agg({
                'coefficient' : 'sum',
            }).reset_index()
#Bảng gồm: username, tổng số lượng assigment mà user đó tham gia
username_sumAssignment = annonimized_df.groupby(['username']).agg({
                'assignment_id' : 'nunique',
            }).reset_index()
username_sumCoef_sumAssigment = pd.merge(username_sumCoef, username_sumAssignment, on = 'username', how = 'inner')
#Tính ra tỉ lệ trễ của user
username_sumCoef_sumAssigment['late_rate'] = abs((username_sumCoef_sumAssigment['coefficient'] / username_sumCoef_sumAssigment['assignment_id'])/100 - 1)
```

```python
#Bảng gồm: username và laterate
username_laterate = username_sumCoef_sumAssigment.drop(columns = ['coefficient', 'assignment_id'])
```

```python
df1 = pd.merge(df, username_problem_df, on = ['username', 'assignment_id'], how = 'inner')
```

```python
df1['finish time'] = (df1['create at'] - df1['begin at']).dt.total_seconds()
```

```python
# Bỏ đi các cột lúc này không còn cần thiết
df1.drop(columns = ['begin at', 'end at', 'create at'], inplace = True)
```

```python
#Bảng gồm: username, assigment_id, số lượng problem tham gia trong assigment, số lượng problem trong assigment, problem_id, max_score (điểm cao nhất mà user đạt được cho problem đó), num of submit (số lần nộp bài cho problem đó), finish time (thời gian để user đó hoàn thành problem)
df1
```

```python
#Bảng này gồm usernam, assignment_id, attend, problem, max_score (tổng điểm của từng problem trong assignment), num of submit (tổng số lần nộp bài trong assignment), finish time (tổng thời gian hoàn thành các problem trong assignment)
df2 = df1.groupby(['username', 'assignment_id', 'attend', 'problem']).agg({
                'max_score' : 'sum',
                'num of submit' : 'sum',
                'finish time' : 'sum'
            }).reset_index()
```

```python
#Bảng gồm: username, attend (tổng số lượng problem mà user đã tham gia). problem (tổng số lượng problem trong assignment mà user đã tham gia), max_score (tổng điểm các assignment), num of submit (tổng số lần nộp bài của tất cả assignment), finish_time (tổng thời gian để hoàn thành tất cả assignment)
df3 = df2.groupby('username').agg({
    'attend' : 'sum',
    'problem' : 'sum',
    'max_score' : 'sum',
    'num of submit' : 'sum',
    'finish time' : 'sum'
}).reset_index()
```

```python
# Tính toán các đặc trưng sẽ sử dụng
df3['attend_rate'] = df3['attend'] / df3['problem']
df3['avg_score'] = (df3['max_score'] / df3['problem']) / 1000
df3['avg_submit'] = df3['num of submit'] / df3['attend']
df3['avg_finishtime (hour)'] = (df3['finish time'] / df3['attend']) / 3600
```

```python
# Bỏ đi các cột lúc này không còn cần thiết
df3.drop(columns = ['problem', 'max_score', 'num of submit', 'finish time'], inplace = True)
```

```python
#Đổi lại tên cột cho df4
df3.columns = ['username','attend',	'attend_rate',	'avg_score',	'avg_submit',	'avg_finishtime (hour)']
#Kết hợp với thuộc tính late_rate từ bảng username_laterate để thu được bảng đầy đủ
df4 = pd.merge(df3, username_laterate, on = 'username' , how = 'inner')
```

```python
# data_candudoan là những user chưa có cột điểm trong các bảng th_df, tbtl_df, ck_df, qt_df
data_candudoan = df4[~df4['username'].isin(tbtl_df['username'])].reset_index()
```

```python
data = pd.merge(df4, tbtl_df, on = 'username', how = 'inner')
```

```python
data
```

```python
data.replace('\xa0', pd.NA, inplace=True)
data.dropna(inplace = True)
```

```python
data.isna().sum()
```

```python
X = data[['attend', 'attend_rate', 'avg_score', 'avg_submit', 'avg_finishtime (hour)', 'late_rate']]
y = data['TBTL']
```

```python
# Khởi tạo tham số cần tìm kiếm
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Khởi tạo mô hình RandomForest
rf = RandomForestRegressor(random_state=42)

# Khởi tạo GridSearchCV
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Huấn luyện mô hình
grid_search_rf.fit(X, y)

# In ra tham số tốt nhất
print(f"Best parameters for RandomForest: {grid_search_rf.best_params_}")
```

```python
best_rf = grid_search_rf.best_estimator_
y = best_rf.predict(data_candudoan[['attend', 'attend_rate', 'avg_score', 'avg_submit', 'avg_finishtime (hour)', 'late_rate']])
```
